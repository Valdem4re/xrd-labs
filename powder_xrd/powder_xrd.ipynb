```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
def read_file(file_name: str):
    theta_2 = []
    intense = []
    with open(file_name, 'r') as file:
        for line in file:
            data = line.split()
            # Преобразуем строки в числа
            theta_2.append(float(data[0]))
            intense.append(float(data[1]))
    return theta_2, intense
```

```python
KDP100_theta, KDP100_I = read_file("KDP100_27112023_clear.txt")
peaks, _ = find_peaks(KDP100_I, height=3000)
KDP100_theta = np.array(KDP100_theta)
KDP100_I = np.array(KDP100_I)
print(KDP100_theta[peaks])
print(KDP100_I[peaks])
plt.plot(KDP100_theta[peaks], KDP100_I[peaks], 'o', label="KDP:100%")

print("peaks length:", len(peaks))
KDP75_theta, KDP75_I = read_file("KDP75_TU25_27112023_clear.txt")

KDP75_theta = np.array(KDP75_theta)
KDP75_I = np.array(KDP75_I)
print(KDP75_theta[peaks])
print(KDP75_I[peaks])
plt.plot(KDP75_theta[peaks], KDP75_I[peaks], 'o', label="KDP:75%")
print("peaks length:", len(peaks))
KDP25_theta, KDP25_I = read_file("KDP25_TU75_27112023.txt")

KDP25_theta = np.array(KDP25_theta)
KDP25_I = np.array(KDP25_I)
print(KDP25_theta[peaks])
print(KDP25_I[peaks])
plt.plot(KDP25_theta[peaks], KDP25_I[peaks], 'o', label="KDP:25%")
plt.legend()
plt.show()
print("peaks length:", len(peaks))
```

```python
avg1 = sum(KDP100_I[peaks]) / len(peaks)
avg2 = sum(KDP75_I[peaks]) / len(peaks)
avg3 = sum(KDP25_I[peaks]) / len(peaks)
i = [avg1, avg2, avg3]
x = ["100%", "75%", "25%"]
plt.xlabel("Концентрация KDP в исследуемом порошке")
plt.ylabel("I")
plt.plot(x, i)
```
